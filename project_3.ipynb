## Data Cleaning

### Data type

```python
df['stop_date'] = pd.to_datetime(df.stop_date, format='%m/%d/%Y')
df.info()
```

```python
df['stop_time'] = pd.to_datetime(df.stop_time, format='%H:%M').dt.time
```

```python
df.head()
```

```python
df.columns
```

### Drop unused columns

```python
df = df.drop(columns=['country_name','driver_age_raw','violation_raw','search_type']).copy()
```

```python
df = df.astype({'driver_gender':'category',\
                'driver_age': 'float16',\
                'driver_race':'category',\
                'violation':'category',\
                'search_conducted':'bool',\
                'stop_outcome':'category',\
                'is_arrested':'bool',\
                'stop_duration':'category',\
                'drugs_related_stop':'bool'
               })
df.info()
```

```python
df.head()
```

```python
df['driver_gender'].isna().sum()
```

```python
df.query('driver_gender.isna() & driver_age.isna() & violation.isna()')
```

### Drop row

```python
print(df.shape)
df = df.drop(df.query('driver_gender.isna() & driver_age.isna() & violation.isna()').index)
print(df.shape)
```

```python
df_arrested = df[df['is_arrested']==True]
```

## Data Exploration in detail: GroupBy

```python
df_arrested.groupby(['driver_race','driver_gender'],observed=False)[['driver_age']].agg(['mean','min','max'])
```

```python
df_arrested.groupby(['search_conducted','violation'],observed=False)[['is_arrested']].count()
```

```python
df_arrested.groupby(['stop_outcome'],observed=False)[['is_arrested']].count().sort_values(by='is_arrested',ascending=False)
```

```python
df_arrested.groupby(['stop_outcome','driver_gender'],observed=False)[['is_arrested']].count().sort_values(by='is_arrested',ascending=False)
```

```python
df_arrested.groupby(['drugs_related_stop'],observed=False)[['is_arrested']].count().sort_values(by='is_arrested')
```

```python
df_arrested_bydate = df_arrested.groupby(['stop_date'],observed=False)[['is_arrested']].count().sort_values(by='is_arrested')
df_arrested_bydate = df_arrested_bydate.sort_index()
```

```python
df_arrested_bydate.info()
```

```python
df_arrested_bydate.hist()
```

```python
df_arrested_bydate.plot(kind='line', figsize = (20,6));
```

```python
df_arrested_bydate.loc['2012-01-01':'2012-06-30', :].plot(kind='line', figsize = (20,6));
```

```python
df_arrested_bydate2 = df_arrested_bydate.reset_index().copy()
df_arrested_bydate2
```

```python
df_arrested_bydate2['year']=df_arrested_bydate2['stop_date'].dt.year
df_arrested_bydate2['month']=df_arrested_bydate2['stop_date'].dt.month
df_arrested_bydate2['day']=df_arrested_bydate2['stop_date'].dt.day
df_arrested_bydate2= df_arrested_bydate2[['stop_date','year','month','day','is_arrested']]
df_arrested_bydate2
```

```python
df_arrested_bydate2.columns
```

```python
ax = df_arrested_bydate2.groupby(['year'],observed=False)[['is_arrested']].count().sort_values(by=['year'],ascending=True)
ax.plot();
```
