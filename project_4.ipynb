```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/khala1391/DSL_Dataset/raw/main/project_4_Covid-19_Dataset.csv', index_col ='Date', parse_dates = True)
df.head(3)
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.astype({'Confirmed': 'int32',
               'Deaths': 'int32',
               'Recovered': 'int32'})
df.info()
```

```python
df[ ~ df.isna().any(axis=1)]
```

```python
df.columns
```

```python
df[ ~ df.isna().any(axis=1)].groupby('Region')[['State']].count()
```

```python
df_big = df[ ~ df.isna().any(axis=1)].groupby('Region')[['Confirmed', 'Deaths', 'Recovered']].sum().sort_values(by='Recovered',ascending=False)
df_big
```

```python
df_big['Recovered_Ratio'] = (df_big['Recovered']/df_big['Confirmed']).copy()
df_big = df_big.sort_values(by='Recovered_Ratio', ascending=False).round(3)
df_big
```

```python
df_big['Recovered_Ratio'].plot(kind='barh', title="Recovered Ratio by Country");
```

# find duplicate by method duplicated (easier)

### check 'State' Column

```python
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset=['State'])]
```

```python
df.loc[df.duplicated(subset=['State'],keep=False)].tail(10)
```

### check 'State' and 'Region' Column >> the most correct way

```python
df.loc[df.duplicated(subset=['State','Region'])].tail(10)
```

- no any duplicated row
