# Multiple Linear Regression

## Importing the dataset

```python
df.columns
```

```python
# df_category_in =['Make', 'Type', 'DriveTrain']
# df_numeric_in = ['EngineSize', 'Cylinders','Horsepower','Weight']
# df_target_out =['MPG_Highway']
# df_selected_cols = df_category_in + df_numeric_in +df_target_out
```

```python
X =df[['DriveTrain','EngineSize', 'Cylinders','Horsepower','Weight']]
y =df[['MPG_Highway']]
```

```python
# X = X.to_numpy()
```

```python
# y = y.to_numpy()
```

```python
# X = df[df_numeric_in]
# y = df[df_target_out]
```

```python
type(X)
```

## Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
# X = np.array(ct.fit_transform(X))
```

```python
X = ct.fit_transform(X)
```

```python
print(X)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
type(X_train)
```

```python
type(X_test)
```

```python
type(y_train)
```

```python
type(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred = np.ravel(y_pred)
```

```python
type(y_pred)
```

```python
y_pred.shape
```

```python
dc=pd.concat([y_test.reset_index(), pd.Series(y_pred, name='Predicted').round(decimals=2)], axis='columns')
dc
```

```python
y_test['MPG_Highway'].max()
```

## Evaluating the Model Performance

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
print('RMSE',sqrt(mean_squared_error(y_test,y_pred)))
```
