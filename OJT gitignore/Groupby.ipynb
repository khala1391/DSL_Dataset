```python
import pandas as pd
%matplotlib inline
from IPython.display import JSON
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df
```

```python
df.describe()
```

```python
df.columns
```

```python
df.groupby('class').describe()
```

```python
df.groupby('class').describe().T
```

```python
g=df.groupby('class')
```

```python
g
```

```python
JSON(g.groups)
```

```python
g.get_group('minivan')
```

```python
g.mean(numeric_only=True)
```

```python
g.mean(numeric_only=True)[['cty', 'hwy']]
```

```python
g.mean(numeric_only=True)[['cty', 'hwy']].sort_values('hwy').plot(kind='barh');
```

```python
h=df.groupby("manufacturer")
```

```python
h
```

```python
JSON(h.groups)
```

```python
h.get_group('honda')
```

```python
h.mean(numeric_only=True)
```

```python
h.mean(numeric_only=True)[['displ','cyl']].sort_values('cyl').plot(kind='barh')
```
